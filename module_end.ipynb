# Detect and handle outliers.

```python
df.columns = df.columns.str.strip()
```

```python
import pandas as pd

# Function to detect and impute outliers using the IQR method
def impute_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Impute outliers with boundary values
        df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
        
    return df

# List of columns to handle outliers (excluding the target column 'Bankrupt?')
columns_to_handle = [
    'Net Income to Total Assets', 'ROA(A) before interest and % after tax',
    'ROA(B) before interest and depreciation after tax', 'ROA(C) before interest and depreciation before interest',
    'Net worth/Assets', 'Debt ratio %', 'Persistent EPS in the Last Four Seasons',
    'Retained Earnings to Total Assets', 'Net profit before tax/Paid-in capital',
    'Per Share Net profit before tax (Yuan ¥)', 'Current Liability to Assets',
    'Working Capital to Total Assets', 'Net Income to Stockholder\'s Equity',
    'Borrowing dependency', 'Current Liability to Current Assets', 'Liability to Equity',
    'Net Value Per Share (A)', 'Net Value Per Share (B)', 'Net Value Per Share (C)',
    'Current Liability to Equity'
]

# Assuming df is your dataframe
df_imputed = impute_outliers_iqr(df.copy(), columns_to_handle)

# Check the result
print(df_imputed.head())
```

# *Perform hypothesis testing to identify significant features influencing bankruptcy.

```python
from scipy.stats import chi2_contingency, ttest_ind

# Example: T-test for numerical features
significant_features = []
for column in df.columns:
    if df[column].dtype in ['int64', 'float64'] and column != 'Bankrupt?':
        stat, p = ttest_ind(df[column][df['Bankrupt?'] == 1],
                            df[column][df['Bankrupt?'] == 0])
        if p < 0.05:
            significant_features.append(column)
            print(f'Feature: {column}, p-value: {p}')
```

# 1.created a empty list
# 2.iterated each column and checked for datatype weather it is int or float type
# 3.performed null hypothesis on each column with respect to 'bankrupt?' column
# 4. compared p value of each column with 0.05 and columns with less than p value than 0.05 are appended in empty list
# 5. this columns which are in the list are the features that influencing bankrupcy

```python
print(significant_features)#the below all columns is significantly influence on bankrupcy
```

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Create Polynomial Features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(df.drop(columns=['Bankrupt?']))
poly_features = pd.DataFrame(X_poly, columns=poly.get_feature_names_out(df.drop(columns=['Bankrupt?']).columns))

# Combine with the target variable
df_poly = pd.concat([poly_features, df['Bankrupt?']], axis=1)
```

# Dimentionaly reducing using PCA

```python
X = df.drop(columns=['Bankrupt?'])
y = df['Bankrupt?']
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X)
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
df_pca['Bankrupt?'] = y
```

# Feature Selection using RFE and RandomForest

```python
# Feature Selection using RFE
model = LogisticRegression(max_iter=1000, random_state=42)
rfe = RFE(model, n_features_to_select=10)
fit = rfe.fit(X, y)
selected_features_rfe = X.columns[fit.support_].tolist()

# Feature Selection using RandomForest
rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)
importances = rf.feature_importances_
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
selected_features_rf = feature_importances.head(10)['Feature'].tolist()
```

```python
X = df_poly.drop(columns=['Bankrupt?'])
y = df_poly['Bankrupt?']
```

# split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

# apply smote to handle imbalanced data

```python
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)
```

```python
# Apply Logistic Regression for classification
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_res_scaled, y_train_res)
```

```python
# Predict on test data
y_pred = log_reg.predict(X_test_scaled)
y_pred_proba = log_reg.predict_proba(X_test_scaled)[:, 1]
```

```python
# Evaluate model performance using appropriate metrics
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
```

# Detailed Report: Key Insights and Recommendations

### 1. Debt Ratio % and Bankruptcy
- Debt Ratio % exhibits the highest positive correlation with bankruptcy, indicating that as this ratio increases, the likelihood of bankruptcy also rises.
- *Actionable Insight:* It is critical to closely monitor and manage the Debt Ratio %. Implementing stringent debt management policies and ensuring that debt remains sustainable relative to assets and income are essential strategies.

### 2. Net Income to Total Assets and Bankruptcy
- Net Income to Total Assets shows the strongest negative correlation with bankruptcy. Lower values of this ratio suggest higher risks of bankruptcy.
- *Actionable Insight:* Prioritize initiatives aimed at improving profitability relative to total assets. Strategies may include optimizing costs, enhancing revenue streams, and maximizing asset efficiency.

### 3. Return on Assets (ROA) and Bankruptcy
- Both ROA(A) and ROA(B) before interest and taxes demonstrate negative correlations with bankruptcy risk. Lower ROA values indicate higher bankruptcy risks.
- *Actionable Insight:* Enhance operational efficiency and profitability metrics. Focus on optimizing resource allocation, refining operational processes, and diversifying revenue sources to mitigate bankruptcy risks.

### 4. Current Liability to Assets
- Higher values of Current Liability to Assets are associated with increased bankruptcy risk, reflecting higher short-term obligations relative to total assets.
- *Actionable Insight:* Effectively manage short-term liabilities by negotiating favorable terms with creditors, optimizing working capital, and ensuring adequate cash flow for debt obligations.

### Recommendations for Financial Risk Management

#### 1. Strengthen Financial Health Monitoring
- Regularly monitor key financial health indicators such as Debt Ratio %, Net Income to Total Assets, and ROA. Establish threshold values and implement alert mechanisms for timely intervention.

#### 2. Develop Robust Risk Mitigation Strategies
- Tailor risk mitigation strategies to identified risk factors. Establish contingency plans for managing debt levels during economic downturns or adverse market conditions.

#### 3. Enhance Financial Modeling and Forecasting
- Utilize advanced financial modeling techniques to forecast critical financial ratios and their impact on bankruptcy risk. Conduct scenario analysis to assess business resilience under varying economic conditions.

#### 4. Diversification and Revenue Management
- Explore diversification strategies to reduce reliance on specific revenue sources or markets. Diversification can help mitigate risks associated with market fluctuations and sector-specific challenges.

#### 5. Foster Continuous Improvement and Adaptation
- Foster a culture of continuous improvement in financial management practices. Regularly review and adapt strategies based on evolving market dynamics, regulatory changes, and internal performance insights.
